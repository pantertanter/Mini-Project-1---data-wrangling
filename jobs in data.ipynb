```python
# remove most empty columns, which are not so informative
df = df.drop(['salary_currency', 'salary'], axis=1)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plotting a histogram of salaries
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='salary_in_usd', bins=30, kde=True)
plt.title('Distribution of Salaries')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.show()

# Plotting a count plot of job categories
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='job_category', order=df['job_category'].value_counts().index)
plt.title('Count of Job Categories')
plt.xlabel('Count')
plt.ylabel('Job Category')
plt.show()

# Plotting a box plot of salaries by experience level
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='experience_level', y='salary_in_usd', order=['Entry-level', 'Mid-level', 'Senior'])
plt.title('Salaries by Experience Level')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.show()
```
